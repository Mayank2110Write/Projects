```python
import pandas as pd
```

```python
dataset = pd.read_csv("/content/Churn_Modelling.csv")
```

```python
dataset.head(3)
```

```python
dataset.drop(columns = ["RowNumber", "CustomerId", "Surname", "Geography", "Gender"], inplace = True)
```

```python
dataset["Exited"].value_counts()
```

```python
x = dataset.iloc[:,:-1]
y = dataset.iloc[:,-1]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

```python
x = pd.DataFrame(ss.fit_transform(x), columns = x.columns)
```

```python
import tensorflow
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
x_train.shape
```

```python
from keras.layers import Dense
from keras.models import Sequential
```

```python
ann = Sequential()
```

```python
ann.add(Dense(6, input_dim = 8, activation = "relu"))
ann.add(Dense(4, input_dim = 8, activation = "relu"))
ann.add(Dense(2, input_dim = 8, activation = "relu"))
ann.add(Dense(1, input_dim = 8, activation = "sigmoid"))
```

```python
ann.compile(optimizer = "adam", loss = "binary_crossentropy", metrics = ["accuracy"])
```

```python
ann.fit(x_train, y_train, batch_size=100, epochs =50)
```

```python
from sklearn.metrics import accuracy_score
```

```python
prd = ann.predict(x_test)
```

```python
prd_data = []
for i in prd:
  if i>0.5:
    prd_data.append(1)
  else:
    prd_data.append(0)
```

```python
tr = ann.predict(x_train)
```

```python
tr_data = []

for i in tr:
  if i>0.5:
    tr_data.append(1)
  else:
    tr_data.append(0)
```

```python
#test
accuracy_score(y_test, prd_data)*100
```

```python
#training

accuracy_score(y_train, tr_data)*100
```
